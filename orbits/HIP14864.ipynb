```python
from __future__ import print_function
import PyAstrOFit.Orbit as Orbit
import numpy as np

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

### Instanciate an orbit object

```python
par=39.8220 #parallax mas
asec= 0.5037 #a [arcsec]
a = asec/(par*1.e-3)        #[A.U.]
e = 0.735
i = 46.4       #[deg]
node = 69.9    #[deg]
w = 266.1      #[deg]
tp = 2459488.55122 #[JD]
P = 35.3 #[yr]
starDistance = 1/(par*1.e-3) #[pc]
```

```python
target = Orbit(semiMajorAxis = a,
                   eccentricity = e,
                   inclinaison = i,
                   longitudeAscendingNode = node,
                   periastron = w,
                   periastronTime = tp,
                   dStar = starDistance,
                   period=P)
```

#### Position at specific observation dates

```python
observation_dates=['2020-08-31 2:00:00']
target.showOrbit(pointOfView = 'earth', 
                   unit = 'arcsec', 
                   figsize = (10,10), 
                   title = [r'a probable orbit',20],
                   addPosition = observation_dates)
```
