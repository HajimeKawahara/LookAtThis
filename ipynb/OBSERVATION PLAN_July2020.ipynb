```python
import matplotlib.pyplot as plt
from astropy.time import Time
from astroplan.plots import plot_sky
import numpy as np
from astroplan import FixedTarget, Observer
import astropy.units as u
```

```python
#from astroplan import download_IERS_A
#download_IERS_A()
```

```python
subaru=Observer.at_site('subaru')
```

```python
def ut(hst):
    return np.mod(hst+10,24)
```

```python
hip93935 = FixedTarget.from_name('HIP 93935')
hip93935_style = {'color': 'k'}

BDm184118 = FixedTarget.from_name('BD-18 4118')
BDm184118_style = {'color': 'red'}

BD172803 = FixedTarget.from_name('BD+17 2803')
BD172803_style = {'color': 'green'}

```

```python
hip84514 = FixedTarget.from_name('HIP 84514')
hip84514_style = {'color': 'blue',"marker":"s"}

hip72102 = FixedTarget.from_name( 'HIP 72102')
hip72102_style = {'color': 'orange',"marker":"s"}

BD354113 = FixedTarget.from_name('BD+35 4113')
BD354113_style = {'color': 'gray',"marker":"s"}
```

```python
hip79124 = FixedTarget.from_name('HIP 79124')
hip79124_style = {'color': 'cyan',"marker":"^"}
```

```python
ut(24)
```

```python
nt = Time.now()
```

```python
#HST=Time(['2020-07-02 22:00:00'])
#observe_time = HST+10*u.hour
observe_time = Time.now()
plot_sky(hip93935, subaru, observe_time, style_kwargs=hip93935_style )
plot_sky(BDm184118, subaru, observe_time, style_kwargs=BDm184118_style )
plot_sky(BD172803, subaru, observe_time, style_kwargs=BD172803_style )
plot_sky(hip84514, subaru, observe_time, style_kwargs=hip84514_style)
plot_sky(hip72102, subaru, observe_time, style_kwargs=hip72102_style)
plot_sky(BD354113, subaru, observe_time, style_kwargs=BD354113_style)
plot_sky(hip79124, subaru, observe_time, style_kwargs=hip79124_style)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
#plt.title("HST="+str(HST[0]))
print(observe_time - 10*u.hour)
plt.savefig("pos.png", bbox_inches="tight", pad_inches=0.0)
```
